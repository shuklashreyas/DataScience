# Question 7

(10 pts) Using a 2-sample t-test**, determine if there is a statistical difference between the price of diamonds with a cut that is considered:
'Good' vs 'Very Good'
'Premium' vs 'Ideal'
**set alpha to 5% i.e. 0.05.

```python
from scipy.stats import ttest_ind

# Prices of diamonds with 'Good' and 'Very Good' cuts
good_prices = diamonds_df[diamonds_df['cut'] == 'Good']['price']
very_good_prices = diamonds_df[diamonds_df['cut'] == 'Very Good']['price']

# Prices of diamonds with 'Premium' and 'Ideal' cuts
premium_prices = diamonds_df[diamonds_df['cut'] == 'Premium']['price']
ideal_prices = diamonds_df[diamonds_df['cut'] == 'Ideal']['price']

# Perform 2-sample t-tests
t_stat_good_vs_very_good, p_value_good_vs_very_good = ttest_ind(good_prices, very_good_prices)
t_stat_premium_vs_ideal, p_value_premium_vs_ideal = ttest_ind(premium_prices, ideal_prices)

print(f"t-statistic (Good vs Very Good): {t_stat_good_vs_very_good}, p-value: {p_value_good_vs_very_good}")
print(f"t-statistic (Premium vs Ideal): {t_stat_premium_vs_ideal}, p-value: {p_value_premium_vs_ideal}")
```


The p-value for the comparison between "Good" and "Very Good" cuts (0.419) is greater than 0.05, indicating that there is no statistically significant difference in the prices between these two cuts. In contrast, the p-value for the comparison between "Premium" and "Ideal" cuts is extremely small (close to 0), far below 0.05, indicating a statistically significant difference in prices between these two cuts. The large t-statistic for "Premium vs Ideal" suggests a substantial difference in their mean prices.

(7 pts) [optional/challenge question] Many statistical test rely on the assumption that data is normally distributed. Indicate which distributions in question 6 were skewed, and research one possible technique* that can be used to handle skewed data. Then perform the following:

Demonstrate the selected technique to correct the skewed data.

Repeat the analysis in question 6 using the corrected data and explain the difference in the distribution with the previous results.

Repeat the statistical tests in question 7. Are there any changes in the statistical significance? Explain the results.

*Helpful tip: Click here to watch a video about correcting skewed data.Links to an external site.

```python
# Apply log transformation to the price data
diamonds_df['log_price'] = np.log(diamonds_df['price'])

# Create a boxplot for the distribution of log-transformed prices based on the quality of the cut
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='log_price', data=diamonds_df)
plt.xlabel('Cut')
plt.ylabel('Log Price')
plt.title('Distribution of Log-Transformed Diamond Prices Based on Cut Quality')
plt.show()

# Perform 2-sample t-tests on log-transformed prices
log_good_prices = diamonds_df[diamonds_df['cut'] == 'Good']['log_price']
log_very_good_prices = diamonds_df[diamonds_df['cut'] == 'Very Good']['log_price']
log_premium_prices = diamonds_df[diamonds_df['cut'] == 'Premium']['log_price']
log_ideal_prices = diamonds_df[diamonds_df['cut'] == 'Ideal']['log_price']

t_stat_log_good_vs_very_good, p_value_log_good_vs_very_good = ttest_ind(log_good_prices, log_very_good_prices)
t_stat_log_premium_vs_ideal, p_value_log_premium_vs_ideal = ttest_ind(log_premium_prices, log_ideal_prices)

print(f"t-statistic (Log Good vs Very Good): {t_stat_log_good_vs_very_good}, p-value: {p_value_log_good_vs_very_good}")
print(f"t-statistic (Log Premium vs Ideal): {t_stat_log_premium_vs_ideal}, p-value: {p_value_log_premium_vs_ideal}")
```

The log-transformed boxplot shows a more compressed distribution of diamond prices for all cuts, reducing the effect of extreme outliers that were prominent in the original data. The spread between the different cuts, particularly for "Ideal" and "Premium," remains visible, but the distribution is now more symmetrical, indicating that the skewness has been corrected. After applying the log transformation, the t-test for "Good" vs "Very Good" cuts resulted in a statistically significant p-value (0.0107), meaning there is a significant difference in prices between these cuts after transforming the data. The t-test for "Premium" vs "Ideal" remains highly significant with an extremely small p-value (close to 0), indicating a consistent difference between the two cuts. The log transformation has improved the normality of the data, leading to more reliable statistical test results, particularly for "Good" vs "Very Good" cuts, where the untransformed data did not show a significant difference.
