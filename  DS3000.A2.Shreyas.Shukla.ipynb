```python
import numpy as np

# Load data from CSV file
data = np.genfromtxt('birthwt.csv', delimiter=',', skip_header=1)
print(f"Data dimensions: {data.shape}")

# Calculate the age range of mothers
ages = data[:, 1]
min_age = int(np.min(ages))
avg_age = int(np.mean(ages))
max_age = int(np.max(ages))
print(f"Mother's age range: Min = {min_age}, Avg = {avg_age}, Max = {max_age}")

# Count the number of mothers who smoked during pregnancy
smokers_count = np.sum(data[:, 4] == 1)
print(f"Mothers who smoked during pregnancy: {smokers_count}")

# Count the number of mothers with a history of hypertension
hypertension_count = np.sum(data[:, 6] == 1)
print(f"Mothers with hypertension: {hypertension_count}")

# Count the number of mothers who had one or more premature births
premature_births_count = np.sum(data[:, 5] >= 1)
print(f"Mothers with 1 or more premature births: {premature_births_count}")

# Convert birth weight from grams to pounds and ounces
def convert_birthweight(grams):
    pounds = grams / 453.59237
    ounces = (pounds - int(pounds)) * 16
    return int(pounds), round(ounces, 2)

# Calculate the average birth weight for babies of mothers with premature births
premature_birth_weights = data[data[:, 5] >= 1][:, 9]
avg_premature_weight = np.mean(premature_birth_weights)
print(f"Avg birth weight for premature birth cases: {avg_premature_weight:.2f} grams")

# Calculate birth weight statistics for babies of non-smokers without hypertension
non_smokers_no_htn = data[(data[:, 4] == 0) & (data[:, 6] == 0)][:, 9]
min_bwt_ns_nh = np.min(non_smokers_no_htn)
avg_bwt_ns_nh = np.mean(non_smokers_no_htn)
max_bwt_ns_nh = np.max(non_smokers_no_htn)
print(f"Birth weight (non-smokers, no hypertension): Min = {min_bwt_ns_nh:.2f}, "
      f"Avg = {avg_bwt_ns_nh:.2f}, Max = {max_bwt_ns_nh:.2f} grams")

# Birth weight stats for babies of mothers who smoked during pregnancy
smokers_bwt = data[data[:, 4] == 1][:, 9]
min_bwt_smokers = np.min(smokers_bwt)
avg_bwt_smokers = np.mean(smokers_bwt)
max_bwt_smokers = np.max(smokers_bwt)
print(f"Birth weight (smokers): Min = {min_bwt_smokers:.2f}, "
      f"Avg = {avg_bwt_smokers:.2f}, Max = {max_bwt_smokers:.2f} grams")

# Birth weight stats for babies of mothers with hypertension
hypertension_bwt = data[data[:, 6] == 1][:, 9]
min_bwt_htn = np.min(hypertension_bwt)
avg_bwt_htn = np.mean(hypertension_bwt)
max_bwt_htn = np.max(hypertension_bwt)
print(f"Birth weight (hypertension): Min = {min_bwt_htn:.2f}, "
      f"Avg = {avg_bwt_htn:.2f}, Max = {max_bwt_htn:.2f} grams")

# Birth weight stats for babies of mothers who smoked and had hypertension
smokers_htn_bwt = data[(data[:, 4] == 1) & (data[:, 6] == 1)][:, 9]
min_bwt_smokers_htn = np.min(smokers_htn_bwt)
avg_bwt_smokers_htn = np.mean(smokers_htn_bwt)
max_bwt_smokers_htn = np.max(smokers_htn_bwt)
print(f"Birth weight (smokers & hypertension): Min = {min_bwt_smokers_htn:.2f}, "
      f"Avg = {avg_bwt_smokers_htn:.2f}, Max = {max_bwt_smokers_htn:.2f} grams")
```
